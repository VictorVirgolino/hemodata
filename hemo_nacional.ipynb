```python
import pandas as pd
import glob
import os

pasta_dados_brutos = "./dados_brutos/"

padrao_arquivos = os.path.join(pasta_dados_brutos, "*.xlsx")

lista_de_arquivos = glob.glob(padrao_arquivos)


if not lista_de_arquivos:
    print(f"Nenhum arquivo .xlsx foi encontrado no caminho: {pasta_dados_brutos}")
else:
    print(f"Arquivos encontrados: {len(lista_de_arquivos)}")
    print(lista_de_arquivos)

    # Lista vazia para armazenar cada DataFrame lido
    lista_de_dataframes = []

    # Loop para ler cada arquivo e adicioná-lo à lista
    for arquivo in lista_de_arquivos:
        print(f"Lendo o arquivo: {arquivo}")
        df_temp = pd.read_excel(arquivo)
        lista_de_dataframes.append(df_temp)

    # Concatena (junta) todos os DataFrames da lista em um único DataFrame
    # ignore_index=True reinicia o índice do DataFrame final (0, 1, 2, ...)
    df_final = pd.concat(lista_de_dataframes, ignore_index=True)


    # --- 3. VERIFICAÇÃO DO RESULTADO ---
    print("\n--- DataFrame Final ---")
    print("DataFrame combinado com sucesso!")

    # Mostra informações gerais sobre o DataFrame final
    print("\nInformações do DataFrame:")
    df_final.info()

    # Mostra as 5 primeiras linhas do DataFrame final
    print("\nPrimeiras 5 linhas:")
    display(df_final.head())

    # Mostra as 5 últimas linhas do DataFrame final
    print("\nÚltimas 5 linhas:")
    display(df_final.tail())

    # Opcional: Salvar o DataFrame combinado em um novo arquivo Excel
    # df_final.to_excel('relatorio_consolidado.xlsx', index=False)
    # print("\nDataFrame consolidado salvo em 'relatorio_consolidado.xlsx'")

 
```
